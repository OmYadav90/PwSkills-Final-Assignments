```python
# import lib for Lungs Cancer
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Load the dataset
data = pd.read_csv('dataset.csv')
```

```python
# Checking Data set
data.sample(5)
```

```python
data.head()
```

```python
data.tail()
```

```python

# Overview of the dataset
print("Overview of the dataset:")
print(data.head())
```

```python

# Check for missing values
print("\nMissing values in each column:")
print(data.isnull().sum())
```

```python

# Basic statistics
print("\nBasic statistics:")
print(data.describe())
```

```python

# Count the number of lung cancer cases
cancer_counts = data['LUNG_CANCER'].value_counts()
print("\nCount of Lung Cancer cases:")
print(cancer_counts)
```

```python

# Visualize the distribution of age
plt.figure(figsize=(10, 6))
sns.histplot(data['AGE'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

# Visualize the relationship between smoking and lung cancer
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='SMOKING', hue='LUNG_CANCER')
plt.title('Smoking Status vs Lung Cancer')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.legend(title='Lung Cancer', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
# Visualize the relationship between age and lung cancer
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='LUNG_CANCER', y='AGE')
plt.title('Age vs Lung Cancer')
plt.xlabel('Lung Cancer')
plt.ylabel('Age')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(15, 8))
# Remove non-numeric columns before calculating correlation
correlation_matrix = data.select_dtypes(include=['number']).corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
